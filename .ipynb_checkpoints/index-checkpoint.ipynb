## DATA PRE-PROCESSING

Let's prepare our data fpr modeling a predictive model that answers the following questions:

1.	Can we predict if an employee will leave using available HR data?

2.	Which features (e.g., overtime, department, satisfaction levels) are most important in predicting attrition?

3.	How well does the model perform (accuracy, precision, recall)?

4.	Can the model be used by HR teams to flag high-risk employees for intervention?


#### 1. drop columns with no relevance to our objective (Data wrangling)

```python
# Drop columns with single values or identifiers
df.drop(columns=['EmployeeNumber', 'Over18', 'StandardHours', 'EmployeeCount'], inplace=True)
```

#### 2. Encode the target variable


```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

#### 3. One-Hot Encoding the categorical variables

```python

# Identify categorical features
cat_cols = df.select_dtypes(include='object').columns

# One-hot encode categorical variables
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

#### 4. Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

# Separate features and target
X = df_encoded.drop('Attrition', axis=1)
y = df_encoded['Attrition']

# Scale numeric features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

#### 5. Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

## MODELING

### TRAIN A BASELINE MODEL

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Train the model
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)

# Predict
y_pred = lr.predict(X_test)

# Evaluate the model
print("\n--- Confusion Matrix ---")
print(confusion_matrix(y_test, y_pred))

print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))

```

#### Suspicious Performance:

100% accuracy is rare in real-world scenarios, especially with imbalanced data (83.9% "No" vs. 16.1% "Yes").

**Likely causes:**

Data Leakage: Test set may contain duplicates from the training set.

Overfitting: Model memorized the training data 

Trivial Test Set: Test data might be too small or non-representative.

#### Which features are the most important

```python
import numpy as np
feature_importance = pd.Series(
    lr.coef_[0], index=X.columns).sort_values(ascending=False)

print("Top 10 Features:")
print(feature_importance.head(10))
```
